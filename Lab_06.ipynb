```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D

#sample dataset (3D data)

data = {
'Feature1': [2, 4, 6, 8, 10],
'Feature2': [1, 3, 5, 7, 9],
'Feature3': [5, 10, 15, 20, 25]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Standardize
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# 3D Visualization (Before PCA)

fig = plt.figure(figsize = (12, 5))
ax1 = fig.add_subplot(121, projection = '3d')
ax1.set_xlabel("Feature1")
ax1.set_ylabel("Feature2")
ax1.set_zlabel("Feature3")
ax1.set_title("Before PCA (3D)")
ax1.legend()
ax1.scatter(df['Feature1'], df['Feature2'], df['Feature3'], color = 'blue', label = 'Original Data')

pca = PCA(n_components = 2)

principal_components = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(data = principal_components, columns = ['PC1', 'PC2'])

# 3D Visualization (After PCA)
ax2 = fig.add_subplot(122, projection = '3d')

ax2.set_xlabel("Principal Component 1")
ax2.set_ylabel("Principal Component 2")
ax2.set_title("After PCA (2D)")
ax2.legend()
ax2.grid()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Subjects on X-axis
subjects = ['Math', 'Science', 'English']

# Assigning colors to subjects
colors = ['red', 'blue', 'green']

y_positions = np.arange(len(subjects))  # Positions for bars

plt.figure(figsize=(6, 4))
plt.bar(subjects, [1, 1, 1], color=colors, width=0.6)

# Labeling the Y-axis with colors
plt.yticks([])  # Remove default Y-ticks
for i, color in enumerate(colors):
    plt.text(-0.5, i, color, fontsize=12, color=color, verticalalignment='center')

plt.xlabel("Subjects")
plt.title("Subjects Represented by Colors")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Subjects and their scores for multiple years
years = ['2021', '2022', '2023', '2024']
subjects = ['Math', 'Science', 'English']
scores = [[80, 90, 70], [70, 80, 90], [90, 70, 80], [100, 75, 50]]  # Example scores
colors = ['red', 'blue', 'green']

x = np.arange(len(years))  # Positions for years
width = 0.2  # Width of bars

plt.figure(figsize=(15, 5))

# Plot bars for each subject
for i, subject in enumerate(subjects):
    plt.bar(x + i * width - width, [scores[j][i] for j in range(len(years))], width=width, label=subject, color=colors[i])

plt.xticks(x, years)
plt.ylabel("Scores")
plt.xlabel("Year")
plt.title("Subject Scores Over the Years")
plt.legend()
plt.show()
```
