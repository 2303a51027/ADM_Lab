# Scatter

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]

plt.scatter(x, y, marker = '*', label = 'Data Points', color = 'red')

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Sample Scatter Plot')


plt.legend()
plt.show()
```

# Bar Graph

```python
import matplotlib.pyplot as plt

classes = ['A', 'B', 'C', 'D', 'E']
students = [50, 40, 45, 35, 60]

plt.bar(classes, students, color = 'green')

plt.xlabel("No : Of students")
plt.ylabel("No : Of classes")
plt.title('No : Of students in Different classes')

plt.show()
```

# Histogram

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)

plt.hist(data, bins = 20, color = 'blue', edgecolor = 'black', alpha = 0.7)

plt.xlabel('No : Of students')
plt.ylabel('No : Of classes')

plt.title('No : Of students in Different classes')

plt.show()
```
